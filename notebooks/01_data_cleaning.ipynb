```python
# Import libraries
import pandas as pd
import numpy as np

# Load the raw data file
file_path = '../data/raw_data.csv'
df = pd.read_csv(file_path)

# Display basic information about the dataset
print("\nðŸ”Ž Initial Dataset Info:")
print(df.info())

# Display the first 5 rows to understand the data structure
print("\nðŸ“Š Initial Dataset Sample:")
print(df.head())

# Display basic statistical details of numerical columns
print("\nðŸ“ˆ Summary Statistics:")
print(df.describe())

# âœ… Step 1: Handle Missing Values
# Fill missing Agent_Rating with mean value and Weather with mode (most frequent value)
df.fillna({'Agent_Rating': df['Agent_Rating'].mean(), 
           'Weather': df['Weather'].mode()[0]}, inplace=True)

# âœ… Step 2: Handle Missing Time Values
# Fill missing Order_Time and Pickup_Time with default '00:00:00'
df.fillna({'Order_Time': '00:00:00', 
           'Pickup_Time': '00:00:00'}, inplace=True)

# âœ… Step 3: Confirm Fixes
print("\nâœ… Updated Dataset Info (After Handling Missing Values):")
print(df.info())

# âœ… Step 4: Display Sample Output After Cleaning
print("\nðŸ“Š Sample After Cleaning:")
print(df.head())
```
